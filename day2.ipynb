```python
# 문자열 인덱싱 : indexing
# 문자열의 index를 이용해서 원하는 문자(글자하나)를 가져오는 것
# index는 각 문자에 부여된 고유의 번호( 0 부터 1씩 증가)
str1 = 'Life is too short, You need Python'
#                 1         2         3
#       0123456789012345678901234567890123
#           3         2         1 
#       4321098765432109876543210987654321-
print(str1)
print(str1[3]) # str1[index]
print(str1[-31])
print(str1[29])
print(str1[-5])
print(len(str1)) # 34
# n
print(str1[33]) # 34-1
#            34      - 1  = 33
print(str1[len(str1) - 1])
# L
print(str1[0])
print(str1[-34])
print(str1[-len(str1)])
#####
# print(str1[len(str1) + 1]) # 34 + 1 = 35 # string index out of range
#Life
print(str1[0])
print(str1[1])
print(str1[2])
print(str1[3])
print(str1[0]+str1[1]+str1[2]+str1[3])
# indexing : 문자열에서 하나의 문자를 가져오는 방식이다.
```

```python
#문자열 슬라이싱 : Slicing
str1 = 'Life is too short, You need Python'
#                 1         2         3
#       0123456789012345678901234567890123
#           3         2         1 
#       4321098765432109876543210987654321-
print(str1)
print(str1[0:4]) # start=0, end = 4, step = 1
print(str1[0 : 0 + 4])
word_size = 4
print(str1[0 : 0 + word_size])
# start=0, end = 11, step = 2
print(str1[0:11:2])
# index 12, word_size =5인 글자를 출력
print(str1[12 : 12 + 5])
# 19번부터 마지막까지 출력
print(str1[33])
print(str1[19:33+1])
print(str1[19:34])
#                 34
print(str1[19:len(str1)])
print(str1[19:])
print(str1[0:4])
print(str1[:4])
print(str1[:]) # 처음부터 끝까지
# short를 출력
print(str1[12:17])
print(str1[-22:-17])
print(str1[-22:17])
print(str1[12:-17])
a = "20230331Rainy"
'''
2023년
03월
31일
Rainy
'''
print(a[:4] + "년")
print(a[4:4+2] + "월")
print(a[6:6+2] + "일")
print(a[8:])
```

```python
# 문자열 포매팅
# 포맷 문자 : %s : string , %d : decimal, %f : float
print("이숭무의 나이는 35살입니다.")
name = '이숭무'
age = 35
print("%s의 나이는 %d살입니다." % (name, age))
print("이숭무의 나이는 %d살입니다."% (age) )
print("이숭무의 나이는 %d살입니다."% age )
str2 = "이숭무의 나이는 %d살입니다."% age
print(str2)
str2 = "이숭무의 나이는 %d살입니다."% 35
print(str2)
```

```python
# 포맷문자를 이용한 포맷팅에서 정렬 : -는 왼쪽정렬, +는 오른쪽 정렬
print("hi")
print("%s" % "hi")
print("%10s" % "hi")
print("%-10s이숭무" % "hi")
```

```python
a = "이숭무의 나이는 20살이고 키는 175.3입니다"
print(a)
name = "이숭무"
age = 20
height = 175.3
print("%s의 나이는 %d살이고 키는 %f입니다" % (name, age, height))
print("%s의 나이는 %d살이고 키는 %5.1f입니다" % (name, age, height))
print("%s의 나이는 %d살이고 키는 %4.1f입니다" % (name, age, height))
print("%s의 나이는 %d살이고 키는 %.1f입니다" % (name, age, height))
print("%s의 나이는 %s살이고 키는 %s입니다" % (name, age, height))
```

```python
# format함수를 이용한 포맷팅 : format()
"이숭무의 나이는 20살이고 키는 175.3입니다"
print("%s의 나이는 %d살이고 키는 %.1f입니다" % (name, age, height))
print("{0}의 나이는 {1}살이고 키는 {2}입니다".format(name, age, height))
#                                                    0    1     2
print("{1}의 나이는 {0}살이고 키는 {2}입니다".format( age,name,height))
#                                                    0   1     2
print("{1}의 나이는 {0}살이고 키는 {2}입니다".format( age,name,175.1))
```

```python
# format함수에 있는 이름 사용하기
print("{1}의 나이는 {0}살이고 키는 {h}입니다".format( age,name,h=175.1))
#                                                    0    1     x
#print("{n}의 나이는 {0}살이고 키는 {1}입니다".format( age,n=name,175.1))
#                                                    0    x      x
print("{n}의 나이는 {0}살이고 키는 {h}입니다".format( age,n=name,h=175.1))
```

```python
# 정렬 : > (오른쪽), <(왼쪽) , ^(가운데)
print("{1}의 나이는 {0}살이고 키는 {h}입니다".format( age,name,h=175.1))
print("{1:>10}의 나이는 {0:>10}살이고 키는 {h:>10}입니다".format( age,name,h=175.1))
print("{1:<10}의 나이는 {0:<10}살이고 키는 {h:<10}입니다".format( age,name,h=175.1))
print("{1:^10}의 나이는 {0:^10}살이고 키는 {h:^10}입니다".format( age,name,h=175.1))
print("{1:<10}의 나이는 {0:>10}살이고 키는 {h:^10}입니다".format( age,name,h=175.1))
print("{1:#<10}의 나이는 {0:$>10}살이고 키는 {h:&^10}입니다".format( age,name,h=175.1))
```

```python
#f 문자 포매팅
print("이숭무의 나이는 20살이고 키는 175.3입니다")
name = "이숭무"
age = 20
height = 175.3
print(f"{name}의 나이는 {age}살이고 키는 {height}입니다")
# 정렬 : >, <, ^
print(f"{name:10}의 나이는 {age:10}살이고 키는 {height:10}입니다")
print(f"{name:>10}의 나이는 {age:>10}살이고 키는 {height:>10}입니다")
print(f"{name:<10}의 나이는 {age:<10}살이고 키는 {height:<10}입니다")
print(f"{name:^10}의 나이는 {age:^10}살이고 키는 {height:^10}입니다")
print(f"{name:$^10}의 나이는 {age:&^10}살이고 키는 {height:#^10}입니다")
print(f"{name:$^10}의 나이는 {age:->10}살이고 키는 {height:#^10}입니다")
```

```python
# 1. 포맷문자 포매팅
result = "%s의 나이는 %d살이고 키는 %.1f입니다" % (name, age, height)
print(result)
# 2. format함수 포매팅
result = "{0}의 나이는 {1}살이고 키는 {2}입니다".format(name, age, height)
print(result)
# 3. f문자 포매팅
result = f"{name}의 나이는 {age}살이고 키는 {height}입니다"
print(result)
```
