```python
# 함수의 4가지유형
# 1. 일반적인 형식 : 매개변수 , 결과값(반환값)
def add(x, y):
    return 2 * x + 3 * y
```

```python
result = add(3, 6)
print(result)
```

```python
# 2.매개변수는 있고 결과값(반환값)이 없음
def add(x, y):
    print(2 * x + 3 * y)
```

```python
add(3, 6)
```

```python
# 3. 매개변수가 없고 반환값이 있는 경우
def add():
    x = 3
    y = 5
    return 2 * x + 3 * y
```

```python
result = add()
print(result)
```
