LangChain -> Agent -> LlamaIndex -> 데이터 검색 -> Gemini의 순으로 실행이 된다.

## LlamaIndex의 핵심 내부구조
문서 -> Document -> Chunk -> Embedding -> Vector Store -> Retriever -> LLM

## LlamaIndex는 특히 다음과 같은 작업에서 강점
- RAG 시스템 구축
- 기업 문서 질의응답
- PDF 기반 챗봇
- 사내 위키 검색
- 데이터베이스 질의 지원
- 로그 분석
- 코드 저장소 기반 질의응답
- 다수의 문서를 대상으로 한 의미 기반 검색

최근 활용 방식<br />
사용자 질문 -> LangGraph (대화 흐름, 상태 관리) -> LlamaIndex(문서 검색, RAG) -> Gemini / GPT / Llama(답변 생성)<br />
LangGraph(또는 LangChain)가 전체 애플리케이션의 흐름을 제어하고, LlamaIndex는 필요한 문서나 데이터를 찾아 LLM에 제공하는 역할을 맡는 구조입니다

```python
# ==========================================================
# LlamaIndex + Gemini 2.5 Flash
# Web RAG Example
# Google Colab
# ==========================================================

# -------------------------------
# 1. 라이브러리 설치
# -------------------------------
!pip -q install -U \
llama-index \
llama-index-readers-web \
llama-index-llms-google-genai \
llama-index-embeddings-google-genai \
google-genai \
trafilatura


# -------------------------------
# 2. API Key 입력
# -------------------------------
GOOGLE_API_KEY = "여기에_API_KEY를_입력하세요"


# -------------------------------
# 3. LlamaIndex 설정
# -------------------------------
from llama_index.core import Settings

from llama_index.llms.google_genai import GoogleGenAI
from llama_index.embeddings.google_genai import GoogleGenAIEmbedding

Settings.llm = GoogleGenAI(
    model="gemini-2.5-flash",
    api_key=GOOGLE_API_KEY
)

Settings.embed_model = GoogleGenAIEmbedding(
    model="gemini-embedding-001",
    api_key=GOOGLE_API_KEY
)


# -------------------------------
# 4. 웹사이트 읽기
# -------------------------------
from llama_index.readers.web import TrafilaturaWebReader

loader = TrafilaturaWebReader()

documents = loader.load_data(
    urls=[
        "https://www.python.org/"
    ]
)

print("="*70)
print("읽어온 Document 개수 :", len(documents))
print("="*70)


# -------------------------------
# 5. Document 확인
# -------------------------------
print(documents[0].text[:1000])


# -------------------------------
# 6. Node 생성
# -------------------------------
from llama_index.core.node_parser import SentenceSplitter

splitter = SentenceSplitter(
    chunk_size=512,
    chunk_overlap=50
)

nodes = splitter.get_nodes_from_documents(documents)

print("="*70)
print("Node 개수 :", len(nodes))
print("="*70)


# -------------------------------
# 7. 첫 번째 Node 확인
# -------------------------------
print(nodes[0].text)


# -------------------------------
# 8. Vector Index 생성
# -------------------------------
from llama_index.core import VectorStoreIndex

index = VectorStoreIndex.from_documents(documents)

print("="*70)
print("Vector Index 생성 완료")
print("="*70)


# -------------------------------
# 9. Retriever 생성
# -------------------------------
retriever = index.as_retriever(
    similarity_top_k=3
)


# -------------------------------
# 10. Retriever 테스트
# -------------------------------
retrieved_nodes = retriever.retrieve(
    "Python은 무엇인가?"
)

print("="*70)
print("검색 결과")
print("="*70)

for i, node in enumerate(retrieved_nodes):

    print(f"\n===== Node {i+1} =====")
    print(node.text[:500])


# -------------------------------
# 11. Query Engine 생성
# -------------------------------
query_engine = index.as_query_engine(
    similarity_top_k=3
)


# -------------------------------
# 12. 질문하기
# -------------------------------
response = query_engine.query(
    "Python의 특징을 5가지 설명해줘."
)

print("="*70)
print("Gemini 답변")
print("="*70)

print(response)


# -------------------------------
# 13. Source Node 확인
# -------------------------------
print("="*70)
print("답변 생성에 사용된 Source")
print("="*70)

for i, node in enumerate(response.source_nodes):

    print(f"\n===== Source {i+1} =====")

    print("Score :", node.score)

    print(node.text[:500])
```

```python
import os
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex, Settings, PromptTemplate
from llama_index.llms.gemini import Gemini
from llama_index.embeddings.gemini import GeminiEmbedding
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

# 1. API 키 설정 (Google AI Studio 발급 키)
os.environ["GOOGLE_API_KEY"] = "your_api_key"

def main():
    # 2. Gemini 모델 설정 (Flash 모델 및 내장 임베딩 사용)
    Settings.llm = Gemini(model="models/gemini-2.5-flash", temperature=0.2)
    Settings.embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-m3")

    # 3. 데이터 로드 및 인덱싱
    print("🔧 사내 장비 매뉴얼 데이터를 로드하고 인덱싱하는 중...")
    documents = SimpleDirectoryReader("data").load_data()
    index = VectorStoreIndex.from_documents(documents)

    # 4. 답변 품질을 높이기 위한 커스텀 프롬프트 템플릿 정의
    # (장비 정비사처럼 전문적이고 안전을 강조하는 어조로 유도)
    qa_prompt_tmpl_str = (
        "아래 제공된 장비 매뉴얼 정보를 바탕으로 작업자의 질문에 친절하고 정확하게 답해 주세요.\n"
        "만약 정보가 부족하다면 억지로 답변을 지어내지 말고 '해당 정보는 매뉴얼에 포함되어 있지 않습니다'라고 답하세요.\n"
        "특히 안전 수칙(보호구, 전원 차단 등)과 관련된 내용이 있다면 답변에서 강력히 강조해 주세요.\n\n"
        "---------------------\n"
        "{context_str}\n"
        "---------------------\n"
        "질문: {query_str}\n"
        "답변 (전문 엔지니어 톤으로):"
    )
    qa_prompt_tmpl = PromptTemplate(qa_prompt_tmpl_str)

    # 5. 쿼리 엔진 생성 및 프롬프트 적용
    query_engine = index.as_query_engine()
    query_engine.update_prompts(
        {"response_synthesizer:text_qa_template": qa_prompt_tmpl}
    )

    # 6. 현장 정비사의 시나리오 질문 테스트
    print("\n==================================================")
    question = "지금 ETCH-200 장비에 E-0402 에러가 떴어. 어떻게 조치해야 하고 부품 번호는 뭐야? 안전 조치도 같이 알려줘."
    print(f"👷 현장 정비사 질문: {question}")
    print("==================================================")

    # gemini-2.5-flash 모델이 검색된 컨텍스트를 분석하여 답변 생성
    response = query_engine.query(question)

    print("\n AI 가이드 답변:")
    print(response)

if __name__ == "__main__":
    main()
```

```python
import os
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex, Settings, PromptTemplate
from llama_index.llms.gemini import Gemini
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

# 1. Gemini API 키 설정
os.environ["GOOGLE_API_KEY"] = "your_api_key"

def main():
    # 2. Gemini 2.5 Flash 및 임베딩 모델 설정
    Settings.llm = Gemini(model="models/gemini-2.5-flash", temperature=0.1)
    Settings.embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-m3")

    # 3. PDF 파일 로드 및 인덱싱
    # SimpleDirectoryReader는 폴더 내의 .pdf 파일을 자동으로 인식하고 파싱합니다.
    print("📄 data 폴더에서 PDF 문서를 읽어와 인덱싱하는 중...")
    documents = SimpleDirectoryReader("data").load_data()
    index = VectorStoreIndex.from_documents(documents)

    # 4. HR(인사팀) 스타일의 프롬프트 템플릿 설정
    hr_prompt_tmpl_str = (
        "당신은 회사의 친절한 인사팀(HR) AI 담당자입니다.\n"
        "사내 규정 PDF 문서에서 추출된 아래 정보를 바탕으로 직원의 질문에 공손하고 명확하게 답변해 주세요.\n"
        "문서에 없는 내용에 대해서는 추측해서 답변하지 말고, '인사팀 담당자에게 직접 문의해 주세요'라고 안내하세요.\n\n"
        "---------------------\n"
        "{context_str}\n"
        "---------------------\n"
        "직원 질문: {query_str}\n"
        "인사팀 AI 답변:"
    )
    hr_prompt_tmpl = PromptTemplate(hr_prompt_tmpl_str)

    # 5. 쿼리 엔진 생성 및 프롬프트 적용
    query_engine = index.as_query_engine()
    query_engine.update_prompts(
        {"response_synthesizer:text_qa_template": hr_prompt_tmpl}
    )

    # 6. 직원의 질문 시나리오 테스트
    print("\n==================================================")
    question = "올해 결혼을 하게 되었는데, 경조 휴가는 며칠이나 쓸 수 있고 지원금도 나오나요? 그리고 운동 학원비도 자기개발비로 신청 가능한지 궁금합니다."
    print(f"👤 직원 질문: {question}")
    print("==================================================")

    # PDF에서 관련 내용을 찾아 Gemini가 종합 답변 생성
    response = query_engine.query(question)

    print("\n HR AI 답변:")
    print(response)

if __name__ == "__main__":
    main()
```
