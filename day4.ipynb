```python
# 리터널 자료형 : 정수 : 10 , 실수 : 10.5 , 문자열 : "",''
#                변수 하나의 값만 저장 가능하다.

# 파이선 자료형
# 리스트 자료형 : [] : 여러타입의 데이터 여러개 저장하기 위해서
# 튜플 자료형 : () : 여러타입의 데이터 여러개 저장하기 위해서
# 딕셔너리 : {} : 키를 이용해서 데이터를 가져오도록 하자.
#                {"키" : "값",..., "키n":"값n"}
# 집합(set) : {1} : 집합자료형 : 하나 이상의 값이 존재해야 한다.
#            교집합, 합집합, 차집합
#            값의 중복을 허용하지 않는다 {1,2,3,4,5,5} => {1,2,3,4,5} 
#            공집합은 없다.
```

```python
dic1 = {"name": "이숭무"} # 딕셔너리
s1 = {"이숭무"} # 집합
print(dic1)
print(s1) # 집합
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
# 교집합 : &
print(s1 & s2)
print(s1.intersection(s2))
# 합집합 : | ( pipe ) 
print(s1 | s2)
print(s1.union(s2))
# 차집합
print(s1 - s2)
print(s1.difference(s2))
print(s2.difference(s1))
```

```python
#  집합자료형 함수
# 요소 추가
s1 = {1,2,3}
s1.add(7)
print(s1)
s1.add('hello')
print(s1)
#s1.add(1,2,3,4)
#s1.add([1,2,3,4]) # cannot use 'list' as a set element
#s1.add({1,2,3,4}) # cannot use 'set' as a set element
# 여러개를 한번에
s1.update([4,5,6])
print(s1)
s1.update({8,9,10})
print(s1)
s1.update((11,12,13))
print(s1)
# 삭제
s1.remove(8)
print(s1)
s1.clear()
print(s1) # set()
```

```python
# 리터널 자료형 : 정수 : 10 , 실수 : 10.5 , 문자열 : "",''
#                부울 : 참 / 거짓
#                변수 하나의 값만 저장 가능하다.
# 파이선 자료형
# 리스트 자료형 : [] : 여러타입의 데이터 여러개 저장하기 위해서
# 튜플 자료형 : () : 여러타입의 데이터 여러개 저장하기 위해서
# 딕셔너리 : {} : 키를 이용해서 데이터를 가져오도록 하자.
#                {"키" : "값",..., "키n":"값n"}
# 집합(set) : {1} : 집합자료형 : 하나 이상의 값이 존재해야 한다.
#            교집합, 합집합, 차집합
#            값의 중복을 허용하지 않는다 {1,2,3,4,5,5} => {1,2,3,4,5} 
#            공집합은 없다.
```

```python
# 참 : True / 거짓 : False
# false는 변수
#a = false : name 'false' is not defined
a =  False
a = "false" # 문자열
# 관계(비교)연산자 : >, <, >=, <=, ==(같다), !=(다르다)
# 결과가 부울이다.
i1 = 10 
i2 = 20 
print(i1 > i2)
#부울 연산자 : 논리연산자 (and , or)
print(True and True) # True
print(True and False) # False
print(True or False) # True
# 관계연산의 결과는 부울 이므로 관계연산과 논리연산을 같이 사용
mat = 40
eng = 60
kor = 80
### 각 과목이 40점이상이고 평균이 60이상이면 합격(True) 아니면 불합격(False)
result = mat >= 40 and eng >= 40 and kor >= 40 and (mat+eng+kor) / 3 >= 60
print(result)
## 윤년은 4년 마다이지만 100년마다 제외된다 하지만 400년마다는 윤년이다.
year = 2200 #"은 윤년일까요? 평년일끼요?"
result = year % 4 == 0 and year % 100 != 0 or year % 400 == 0
print(result)
if result :
    print("윤년")
else:
    print("평년")
```
