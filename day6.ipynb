```python
# parameter이름을 이용해서 값을 전달
def calc(x1, x2, opt):
  if opt == "add":
    result = x1 + x2
  elif opt == "sub":
    result = x1 - x2
  elif opt == "mul":
    result = x1 * x2
  else :
    result = x1 / x2
  return result
```

```python
y = calc(3, 5, "div")
print(y)
y = calc(3, 5, opt="div")
print(y)
y = calc(opt="div", x1=3, x2=5 ) # parameter이름을 사용하는 경우에는 순서 필요없다
print(y)
y = calc(3, opt="div" , x2=5 )
print(y)
#y = calc(3, x2=5, "div") # parameter이름을 중간에 사용할 수 없다. 오른쪽에서 완쪽으로순차적으로 사용해야 한다. 순서X
```

```python
# 클래스변수 초기화
class Calculator:
    def setData(self, first, second):
        self.first = first
        self.second = second
    def add(self):
        self.result = self.first + self.second
    def div(self):
        self.result = self.first / self.second
```

```python
a = Calculator() # 객체 생성
a.setData(10, 20)
print(a.first)
print(a.second)
a.add()
print(a.result)
a.div()
print(a.result)
```

```python
# 클래스변수 초기화
class Calculator:
    # 생성자
    def __init__(self, first, second): # 객체 생성시 자동으로 실행되는 메서드
        self.first = first
        self.second = second
    def add(self):
        self.result = self.first + self.second
    def div(self):
        self.result = self.first / self.second
```

```python
a = Calculator(10, 20) # 객체 생성
print(a.first)
print(a.second)
a.add()
print(a.result)
a.div()
print(a.result)
```

```python
## 상속 : 다른 클래스로 부터 기능을 받아오는 것
class FourCal(Calculator):
    # def __init__(self, first, second): # 객체 생성시 자동으로 실행되는 메서드
    #     self.first = first
    #     self.second = second
    # def add(self):
    #     self.result = self.first + self.second
    # def div(self):
    #     self.result = self.first / self.second
    def mul(self):
        self.result = self.first * self.second
    def sub(self):
        self.result = self.first - self.second
```

```python
a = FourCal(10, 20)
a.add()
print(a.result)
a.div()
print(a.result)
a.mul()
print(a.result)
```

```python
# 오버라이딩
a = FourCal(0, 15) # self.result : 0 / 15  = 0
a.div()
print(a.result)
```

```python
a = FourCal(15, 0) # self.result : 15 / 0  = 오류
a.div()
print(a.result)
```

```python
# 상속해준 class : 부모클래스 (Calculator)
# 상속 받은  class : 자식클래스(FourCal)
class FourCal(Calculator):
    def mul(self):
        self.result = self.first * self.second
    def sub(self):
        self.result = self.first - self.second
    # 오버라이딩 : 부모클래스로 부터 상속 받은 메서드를 자식클래스에서 다시 정의해서 사용하는 것
    def div(self):
        if self.second == 0:
            self.result = 0
        else:
            self.result = self.first / self.second
```

```python
a = FourCal(15, 0) # self.result : 15 / 0  = 오류
a.div()
print(a.result)
```
